the code was run but the output is irrelavant. time to rerun the experiments without the zero weights \
we will take top 100 of both, and use them together.

```python
import experiment
import pandas as pd

clf, x_cols = experiment.experiment("LogisticRegressionWithoutDataLeakage", False, False, True)
clfXg, x_colsXg = experiment.experimentXgBoost("XGBoostWithoutDataLeakage", False, False, True)
selected_features = experiment.combine(x_cols, clf.coef_[0], clfXg.feature_importances_)
df = pd.DataFrame({'final_field_name':selected_features})
df.to_csv('csv/keep_fields.csv', index=False)
df
```

```python
import shutil

# Define the source file path
source_file = "logs/XGBoostLimitedFeatures_cleaner_list.npy"

# Loop over different values of n_estimators and max_depth
for n_estimators in [60,80,100,120,140]:
    for max_depth in [3,5,7,10,12]:
        # Define the destination file path based on the current values of n_estimators and max_depth
        destination_file = f"logs/xgboost.{n_estimators}.{max_depth}_cleaner_list.npy"
        
        # Perform the file copy operation
        shutil.copy(source_file, destination_file)
        print(f"Copied '{source_file}' to '{destination_file}'")
```

```python
from matplotlib import pyplot as plt
import experiment


n_iter_no_change = 5
validation_fraction = 0.1
ax = plt.figure("f1")
plt.title('f1 across estimators')
for n_estimators in [60,80,100,120,140]:
    for max_depth in [3,5,7,10, 12]:
        algo_parms = {
            "n_iter_no_change":n_iter_no_change,
            "validation_fraction": validation_fraction,
            "max_depth":max_depth,
            "n_estimators":n_estimators}
        print('starting experiment', n_estimators, max_depth)
        experiment.experimentXgBoost('xgboost.'+str(n_estimators)+'.'+str(max_depth), False, False, True, algo_parms, ax)
plt.ylabel('f1')
plt.xlabel('n_estimators')
ax.show()
```
